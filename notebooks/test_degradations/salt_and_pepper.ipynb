## Test Salt and Pepper Degradation
Rumore sale e pepe applicato a un'immagine a colori 

```python
from src.degradations.salt_and_pepper import salt_and_pepper_noise
import numpy as np
import cv2
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
# Load an example image
image_path = Path("../../data/raw/DIV2K_train_HR/0010.png")

if not image_path.exists():
    raise FileNotFoundError(f"Image not found at {image_path}")

img = cv2.imread(str(image_path))
if img is None:
    raise ValueError(f"Failed to load image from {image_path}")

# Convert BGR to RGB
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

print(f"✅ Image loaded successfully from {image_path}")
print(f"Image shape: {img.shape}")
print(f"Image dtype: {img.dtype}")
print(f"Color range: {img.min()} - {img.max()}")
```

```python
# Apply salt and pepper noise

noisy_img = salt_and_pepper_noise(img, amount=0.5, salt_vs_pepper=0.5)

# Display the original and noisy images
plt.figure(figsize=(12, 6))
plt.subplot(121)
plt.title("Original Image")
plt.imshow(img)
plt.axis('off')
plt.subplot(122)
plt.title("50% Salt and Pepper Noise, 50/50 distribution")
plt.imshow(noisy_img)
plt.axis('off')
plt.show()
```

```python
# Generate images with only salt noise and only pepper noise

salty_img = salt_and_pepper_noise(img, amount=0.3, salt_vs_pepper=1)
peppery_img = salt_and_pepper_noise(img, amount=0.3, salt_vs_pepper=0)

# Display the salty and peppery images
plt.figure(figsize=(12, 6))
plt.subplot(131)
plt.title("Original Image")
plt.imshow(img)
plt.axis('off')

plt.subplot(132)
plt.title("30% Salt Noise")
plt.imshow(salty_img)
plt.axis('off')

plt.subplot(133)
plt.title("30% Pepper Noise")
plt.imshow(peppery_img)
plt.axis('off')

plt.show()
```
